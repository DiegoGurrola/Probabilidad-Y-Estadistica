```python
from statistics import *
#Lista de calificaciones del grupo a de probabilidad
grupoa= [70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
#la funcion type nos devuelve el tipo de dato de la variable
type(grupoa)
#sacar la media(promedio) de las calificacion
promedio=sum(grupoa)/len(grupoa)
print(promedio)
#utilizar la funcion mean
#el promedio (sgunda opción)
promedio2=mean(grupoa)
print(promedio2)
#valor de la mediana con la funcion median
mediana=median(grupoa)
print(mediana)
#multiples modas que puedan existir, en caso de que exista mas de una moda
moda2=multimode(grupoa)
print(moda2)
#tiempos que tardan los 10 alumnos para correr los 100m
#tiempo en que tardaron los 10 alumnos para recorrer los 100m(promedio o media)
tiempos100mts=[15.10,17.20,14.69,13.27,22.15,18.71,19.5,20,15.10,17.20]
print(tiempos100mts)
print("el tiempo promedio de los corredores ",mean(tiempos100mts))
#median
mediana=median(tiempos100mts)
print("la mediana de los tiempos de los corredores es",mediana)
#mode
moda=mode(tiempos100mts)
print("la moda de los tiempos es=>",moda)
#multimoda
multimodas=multimode(tiempos100mts)
print("la moda de los tiempos es =>",multimodas)
#libreria para generar numeros aleatorios
import random
fabrica1=[]
for i in range(100):
    numero_aleatorio = round(random.uniform(50,100),2)
    fabrica1.append(numero_aleatorio)
print(fabrica1)
#Fabrica
fabrica2=[]
for i in range(100):
    numero_aleatorio =round(random.uniform(40,93),2)
    fabrica2.append(numero_aleatorio)
print(fabrica2)
#libreria python para graficar valores
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.title("Producción diaria de la fabrica 1")
#produccion diaria de la fabrica 2
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
#fab 1
media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)

print("Valores de la fabrica 1")
print("la media es =",media1)
print("la mediana es=",mediana1)
print("la moda es =",moda1)
#fab 2
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)

print("Valores de la fabrica 2")
print("la media es =",media2)
print("la mediana es=",mediana2)
print("la moda es =",moda2)

#HISTOGRAMAS
#generar2 subplot de 1 renglon por 2 columnas para desplegar los 2 histogramas,una al lado del la otra
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
#linea de colores
plt.hist(fabrica1,bins=10,alpha=0.2,color="b", edgecolor="black")
plt.axvline(media1,color="b",linestyle="dashed",linewidth=2,label="media")
plt.axvline(mediana1,color="r",linestyle="dashed",linewidth=2,label="mediana")
plt.axvline(moda1,color="y",linestyle="dashed",linewidth=2,label="moda")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title(" Produccion diaria de la fabrica 1")
plt.legend()
plt.show()

plt.hist(fabrica2,bins=10,alpha=0.2,color="b", edgecolor="black")
plt.axvline(media2,color="b",linestyle="dashed",linewidth=2,label="media2")
plt.axvline(mediana2,color="r",linestyle="dashed",linewidth=2,label="mediana2")
plt.axvline(moda2,color="y",linestyle="dashed",linewidth=2,label="moda2")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title(" Produccion diaria de la fabrica 2")
plt.legend()
plt.show()
```
