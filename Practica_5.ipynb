**Practica 5:  Estadística descriptiva Dataset Autos**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("autos2.csv")
print(df.head())

print(df['price'].describe())
print("Valor mínimo de precios:", df['price'].min())
print("Valor máximo de precios:", df['price'].max())
print("Moda de precios:", df['price'].mode()[0])
print("Media de precios:", df['price'].mean())
print("Mediana de precios:", df['price'].median())
print("Desviación estándar de precios:", df['price'].std())

n = len(df['price'])
bins = int(1 + 3.322 * np.log10(n))
print("Cantidad óptima de bins (Sturges):", bins)

plt.figure(figsize=(10,6))
plt.hist(df['price'], bins=bins, color="skyblue", edgecolor="black")
plt.title("Histograma de Precios de Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x="body-style", y="price", data=df, palette="Set2")
plt.title("Diagrama de Caja: Precio vs Tipo de Vehículo (Body Style)")
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x="drive-wheels", y="price", data=df, palette="Set3")
plt.title("Diagrama de Caja: Precio vs Tipo de Tracción (Drive Wheels)")
plt.xlabel("Tipo de Tracción")
plt.ylabel("Precio")
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x="engine-location", y="price", data=df, palette="coolwarm")
plt.title("Diagrama de Caja: Precio vs Ubicación del Motor (Engine Location)")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()
```

**Estadísticas descriptivas**

Los precios presentan un mínimo de {min_price} y un máximo de {max_price}, lo que muestra un rango amplio entre vehículos económicos y más costosos.  
La media (~{mean_price}) y la mediana (~{median_price}) son relativamente cercanas, lo cual indica que la distribución no está tan sesgada.  
La moda ({mode_price}) sugiere que cierto grupo de autos se repite en el rango de precios. La desviación estándar ({std_price}) muestra que existe una variabilidad considerable.

**Histograma de precios**

El histograma revela cómo se distribuyen los precios. La mayoría de los autos se concentran en un rango bajo-medio, lo que significa que los vehículos más caros son menos frecuentes.  
El número de bins fue calculado con la Regla de Sturges, asegurando una representación estadísticamente adecuada.

**Boxplot Precio vs Body Style**

El gráfico muestra cómo el tipo de carrocería (body style) influye en el precio.  
Algunas carrocerías presentan medianas más altas, lo que indica que ciertos estilos tienden a ser más costosos. También se identifican outliers que corresponden a vehículos fuera del rango típico.

**Boxplot Precio vs Drive Wheels**

Se observa que los autos con distintos tipos de tracción muestran variaciones claras en precios.  
Algunos tipos de tracción están asociados a precios más altos, lo que refleja diferencias tecnológicas y de desempeño.

**Boxplot Precio vs Engine Location**

La ubicación del motor también influye en el precio. Los vehículos con configuraciones menos comunes tienden a tener mayor valor.  
Esto refleja cómo decisiones de diseño impactan directamente en el costo final del automóvil.

**Conclusiones**

El análisis realizado muestra que los precios de los automóviles presentan una distribución amplia, con valores extremos que representan autos fuera de lo común.  
Las estadísticas descriptivas permiten identificar las medidas de tendencia central y variabilidad, mientras que los histogramas y diagramas de caja muestran cómo factores como la carrocería, la tracción y la ubicación del motor influyen en el precio.  
En conclusión, el estudio confirma que las características técnicas y de diseño del vehículo tienen un impacto directo en su valor, y que los datos pueden usarse para tomar decisiones informadas en la industria automotriz.
