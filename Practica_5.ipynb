**Practica 5:  Estad√≠stica descriptiva Dataset Autos**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("autos2.csv")
print(df.head())

print(df['price'].describe())
print("Valor m√≠nimo de precios:", df['price'].min())
print("Valor m√°ximo de precios:", df['price'].max())
print("Moda de precios:", df['price'].mode()[0])
print("Media de precios:", df['price'].mean())
print("Mediana de precios:", df['price'].median())
print("Desviaci√≥n est√°ndar de precios:", df['price'].std())

n = len(df['price'])
bins = int(1 + 3.322 * np.log10(n))
print("Cantidad √≥ptima de bins (Sturges):", bins)

plt.figure(figsize=(10,6))
plt.hist(df['price'], bins=bins, color="skyblue", edgecolor="black")
plt.title("Histograma de Precios de Autom√≥viles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x="body-style", y="price", data=df, palette="Set2")
plt.title("Diagrama de Caja: Precio vs Tipo de Veh√≠culo (Body Style)")
plt.xlabel("Tipo de Veh√≠culo")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x="drive-wheels", y="price", data=df, palette="Set3")
plt.title("Diagrama de Caja: Precio vs Tipo de Tracci√≥n (Drive Wheels)")
plt.xlabel("Tipo de Tracci√≥n")
plt.ylabel("Precio")
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x="engine-location", y="price", data=df, palette="coolwarm")
plt.title("Diagrama de Caja: Precio vs Ubicaci√≥n del Motor (Engine Location)")
plt.xlabel("Ubicaci√≥n del Motor")
plt.ylabel("Precio")
plt.show()
```

**Estad√≠sticas descriptivas**

Los precios presentan un m√≠nimo de {min_price} y un m√°ximo de {max_price}, lo que muestra un rango amplio entre veh√≠culos econ√≥micos y m√°s costosos.  
La media (~{mean_price}) y la mediana (~{median_price}) son relativamente cercanas, lo cual indica que la distribuci√≥n no est√° tan sesgada.  
La moda ({mode_price}) sugiere que cierto grupo de autos se repite en el rango de precios. La desviaci√≥n est√°ndar ({std_price}) muestra que existe una variabilidad considerable.

**Histograma de precios**

El histograma revela c√≥mo se distribuyen los precios. La mayor√≠a de los autos se concentran en un rango bajo-medio, lo que significa que los veh√≠culos m√°s caros son menos frecuentes.  
El n√∫mero de bins fue calculado con la Regla de Sturges, asegurando una representaci√≥n estad√≠sticamente adecuada.

**Boxplot Precio vs Body Style**

El gr√°fico muestra c√≥mo el tipo de carrocer√≠a (body style) influye en el precio.  
Algunas carrocer√≠as presentan medianas m√°s altas, lo que indica que ciertos estilos tienden a ser m√°s costosos. Tambi√©n se identifican outliers que corresponden a veh√≠culos fuera del rango t√≠pico.

**Boxplot Precio vs Drive Wheels**

Se observa que los autos con distintos tipos de tracci√≥n muestran variaciones claras en precios.  
Algunos tipos de tracci√≥n est√°n asociados a precios m√°s altos, lo que refleja diferencias tecnol√≥gicas y de desempe√±o.

**Boxplot Precio vs Engine Location**

La ubicaci√≥n del motor tambi√©n influye en el precio. Los veh√≠culos con configuraciones menos comunes tienden a tener mayor valor.  
Esto refleja c√≥mo decisiones de dise√±o impactan directamente en el costo final del autom√≥vil.

**Conclusiones**
