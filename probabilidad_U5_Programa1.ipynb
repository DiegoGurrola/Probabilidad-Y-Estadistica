```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import google.colab.files # Replaced tkinter with google.colab.files

plt.close('all')

def cargar_archivo():
    # Use google.colab.files.upload() for file selection in Colab
    uploaded = google.colab.files.upload()
    if uploaded:
        # Return the filename of the first uploaded file
        filename = list(uploaded.keys())[0]
        return filename
    else:
        return None # Handle case where no file is selected

def procesar_grupo(ruta, nombre):
    if ruta is None:
        print(f"No se seleccionó ningún archivo para el {nombre}. Saltando procesamiento.\n")
        return

    df = pd.read_csv(ruta)

    if "HorasEstudio" not in df.columns or "Calificacion" not in df.columns:
        print(f"ERROR: El archivo del {nombre} no contiene las columnas necesarias (HorasEstudio, Calificacion).\n")
        return

    X = df["HorasEstudio"].values.reshape(-1, 1)
    y = df["Calificacion"].values

    modelo = LinearRegression()
    modelo.fit(X, y)

    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    r2 = modelo.score(X, y)

    print("\n==============================")
    print(f"RESULTADOS {nombre}")
    print("==============================")
    print(f"Coeficiente de correlación (R²): {r2:.4f}")
    print(f"Ecuación de regresión: y = {pendiente:.4f}x + {intercepto:.4f}")

    if pendiente > 0:
        interpretacion = "Existe una relación lineal **positiva**: a más horas de estudio, mayor calificación."
    elif pendiente < 0:
        interpretacion = "Existe una relación lineal **negativa**: a más horas de estudio, menor calificación."
    else:
        interpretacion = "No existe relación lineal: las horas de estudio no afectan la calificación."

    print("Interpretación:", interpretacion)
    print("======================================================\n")

    plt.figure()
    plt.scatter(df["HorasEstudio"], df["Calificacion"])
    plt.plot(df["HorasEstudio"], modelo.predict(X))
    plt.xlabel("Horas de Estudio")
    plt.ylabel("Calificación")
    plt.title(f"Regresión Lineal - {nombre}")
    plt.show()



# CARGA DE ARCHIVOS

print("Selecciona el archivo del GRUPO 1")
ruta1 = cargar_archivo()
procesar_grupo(ruta1, "GRUPO 1")

print("Selecciona el archivo del GRUPO 2")
ruta2 = cargar_archivo()
procesar_grupo(ruta2, "GRUPO 2")

print("Selecciona el archivo del GRUPO 3")
ruta3 = cargar_archivo()
procesar_grupo(ruta3, "GRUPO 3")
```

1. Interpretación del coeficiente de regresión

Conclusión del Grupo 1
