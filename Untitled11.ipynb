**Practica 1: Revision de pesos**

```python
from scipy.stats import norm

media = 70
desviacion = 4
n = 500

x1 = 68
x2 = 72

p_x1 = norm.cdf(x1, media, desviacion)
p_x2 = norm.cdf(x2, media, desviacion)

probabilidad = p_x2 - p_x1
estudiantes = probabilidad * n

print(f"P(68 < X < 72) = {probabilidad:.4f}")
print(f"Estudiantes que pesan entre 68 y 72 kg: {estudiantes:.0f}")

```
